```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import numpy as np
```

```python
path = './input/'
filename = path + 'vint_single_trace.txt'
```

```python
f = open(filename)
lines = f.read().splitlines()
f.close()
for i, line in enumerate(lines):
    if i < 10:
        print(i, line)
```

```python
df_raw = pd.DataFrame()
```

```python
column_names = ['TWT', 'Vp']
```

```python
rowstoskip = 2
sample_rate = 24
```

```python
df_raw = pd.read_csv(filename, names=column_names, skiprows=rowstoskip, delim_whitespace=True)
```

```python
df_raw['delta TWT'] = sample_rate
```

```python
df_raw['delta Depth'] = df_raw['Vp'] * df_raw['delta TWT'] / 2000
```

```python
df_raw['Depth'] = df_raw['delta Depth'].cumsum(axis=0)
```

```python
df_raw
```

```python
df_raw.plot(x='Vp', y=['TWT', 'Depth'], figsize=(16, 12), grid=True, ylim=(0, 5000), xlim=(0, 6000) )
#df_raw.plot(x='Vp', y='Depth', figsize=(16, 12), grid=True)
```
